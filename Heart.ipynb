```python
import pandas as pd
import numpy as np
import math
df=pd.read_csv('framingham.csv')
df.head()
```

```python
df.info()
```

```python
series=pd.isnull(df['cigsPerDay'])
series
```

```python
data=df.drop(['currentSmoker','education'],axis=1)
cigs=data['cigsPerDay']
cig_mean=cigs.mean()
int_val=math.floor(cig_mean)
cigs.fillna(int_val,inplace=True)
data.dropna(axis=0,inplace=True)
```

```python
cigs
```

```python
data.info()
```

```python
hight_risk=data[data.TenYearCHD==1]
low_risk=data[data.TenYearCHD==0]
final=data.drop(['diaBP','BMI','heartRate'],axis=1)
final
```

```python
X=final[['male','age','cigsPerDay','BPMeds','prevalentHyp','diabetes','totChol','sysBP','glucose']]
y=final['TenYearCHD']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.20,random_state=99)
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(x_train,y_train)
model.score(x_test,y_test)*100
```
